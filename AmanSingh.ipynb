Given two matrices (n * k) and (k * m) , their multiply will give a (n*m) matrix.
For each cell in result matrix we need to do k multiply and k-1 addition. 
Total multiplication needed = n * m * k .
Total addition needed       = n * m * (k-1)

### Answer 2

Given matrix A(3*11) and B(11*3) find their product AB and comare time required for normal python and using numpy

```python
import random 
A = [[random.randint(1, 10) for _ in range(11)] for _ in range(3)]
B = [[random.randint(1, 10) for _ in range(3)] for _ in range(11)]
AB = [[0 for _ in range(3)] for _ in range(3)]
```

```python
def normal(A,B,AB):
    for a in range(len(A)): 
        for b in range(len(B[0])): 
            for k in range(len(B)): 
                AB[a][b] = AB[a][b] + A[a][k] * B[k][b]
normal(A,B,AB)
print(AB)
%timeit normal(A,B,AB)
AB = [[0 for _ in range(3)] for _ in range(3)]
```

```python
import numpy as np
AB = np.dot(A,B)
print(AB)
%timeit np.dot(A,B)
```

mean time for normal python  = 6.2 µs ± 7.97 ns.     
mean time using numpy funtion = 5.28 µs ± 116 ns.

### Answer 3

1)Basic method will give N^2 complexity since you have travel the whole array atleast N/2 time 

2) Better method will be to sort the array and find the median by running single time which take NlogN time complexity

3) By using numpy.median 

```python
import numpy as np 
import time
```

```python
#using 1st method
arr = [5,3,2,1,7,8,9,6,4,11,10]
start = time.time()
print(arr)
for i in range(6):
    mx = 0
    mxind = -1
    for j in range(i,11):
        if(arr[j]>mx):
            mx = arr[j]
            mxind = j
    temp = arr[mxind]
    arr[mxind] = arr[i]
    arr[i] = temp

print(f'median = {arr[5]}')
print(f'time of 1st method = {time.time()- start}')
```

```python
#using 2nd method
arr2 = [5,3,2,1,7,8,9,6,4,11,10]
start = time.time()
arr2 = np.array(arr)
arr2 = np.sort(arr2)
print(arr2)
print(f'median = {arr2[5]}')
print(f'time of 2nd method = {time.time()- start}')
```

```python
#using 3rd method 
arr = [5,3,2,1,7,8,9,6,4,11,10]
start = time.time()
print(f'median = {np.median(arr)}')
print(arr)
print(f'time of 3rd method = {time.time()- start}')
```

### Answer 5

```python

import jax
import jax.numpy as jnp

def my_function(x, y):
    return 2 * x * y + y**3 * jnp.cos(x)

# Define the gradient function using JAX's autograd
grad_my_function = jax.grad(my_function, argnums=(0, 1))

# Choose random values for x and y
random_x = 1.5
random_y = 2.0

# Evaluate the gradient using JAX
jax_gradient = grad_my_function(random_x, random_y)

# Calculate the analytical gradient
analytical_gradient_x = 2 * random_y - random_y**3 * jnp.sin(random_x)
analytical_gradient_y = 2 * random_x + 3 * random_y**2 * jnp.cos(random_x)

# Compare the results
print("JAX Gradient:", jax_gradient)
print("Analytical Gradient:", (analytical_gradient_x, analytical_gradient_y))
```

```python
import sympy as sp

# Define symbolic variables
x, y = sp.symbols('x y')

# Define the function
my_function_sympy = x**2 * y + y**3 * sp.sin(x)

# Calculate the partial derivatives
gradient_x = sp.diff(my_function_sympy, x)
gradient_y = sp.diff(my_function_sympy, y)

# Print the symbolic expressions for the partial derivatives
print("Gradient with respect to x:", gradient_x)
print("Gradient with respect to y:", gradient_y)
```

### Answer 12

```python
import numpy as np

# Creating a 1D array
x = np.array([1, 2, 3])

# Broadcasting to a 2D array
y = x[:, np.newaxis]  # This adds a new axis along the second dimension

# Creating a 2D array
z = np.array([[10, 20, 30],
              [40, 50, 60]])

# Performing element-wise addition using broadcasting
result = x + z

# Printing the arrays and the result
print("Array x:")
print(x)

print("\nArray y (after broadcasting):")
print(y)

print("\nArray z:")
print(z)

print("\nResult of element-wise addition using broadcasting:")
print(result)
```

In cases where broadcasting is not possible, we may need to reshape our arrays or explicitly handle the dimensions to make them compatible for the desired operation.
